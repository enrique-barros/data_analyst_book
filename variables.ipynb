# Variables

# Declaración e inicialización de variable en Python

```python
nombre = "Enrique"

print(nombre)
```

# Reasignación en variables



```python
nombre = "Enrique"
nombre = "Quique"

print(nombre)
```

# Operación aritmética en una variable

```python
suma = 10 + 5

print(suma)
```

# Mostrar resultados con mensajes descriptivos

```python
edad = 34
nombre = "Enrique"

print("Nombre:", nombre)
print("Edad:", edad)
```

# Concatenación

```python
nombre = "Enrique"

print("Nombre del autor: " + nombre)
```

# Mostrar resultados con f-strings

```python
nombre = "Enrique"
edad = 34

print(f"Nombre: {nombre}, Edad: {edad}")
```
